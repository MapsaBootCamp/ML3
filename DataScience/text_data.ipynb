### Regex

A regex is a special sequence of characters that defines a pattern for complex string-matching functionality.

<img src="regex_metacharacters.png">

```python
import re
```

### re.match(pattern, string, flags=0)
The re.match function returns a match object on success, None on failure. We use group(num) or groups() function of match object to get matched expression.

```python
text = "as boot camp ml mapsa"
match = re.match(".*(boot)?(Camp)", text, re.I)

if match:
    print("match: ", match.groups())
else:
    print("no match")
```

### re.search(pattern, string, flags=0)
The re.search function returns a match object on success, none on failure. We use group(num) or groups() function of match object to get matched expression.

```python
text = "as9121232123a:            +98912-123-1234 b: 91254321234 c: 09121234123 d: 912342 e: 09351234521"
pattern = r"(\s+|^)((\+98)|0)?9\d{2}-?\d{3}-?\d{4}(\D|$)"

match = re.search(pattern, text, re.I)

if match:
    print("match: ", match.group())
else:
    print("no match")
```

### re.sub(pattern, repl, string, max=0)
This method replaces all occurrences of the RE pattern in string with repl, substituting all occurrences unless max provided. This method returns modified string.

```python
text = "as boot camp ml mapsa as boot camp"
match = re.sub("boot ?camp", "replaced", text, 2)

if match:
    print(match)
else:
    print("no match")
```

```python
email = "email: ashkan.map@gmail.com, email: foad@gmail.com" 
pattern = "(([a-zA-Z_]+\.?\w+)@([a-zA-Z]+\w+)\.([a-zA-Z]{2,3}))"

match = re.findall(pattern, email)
if match:
    print(match)
```

```python
sample_text = """
                Yann André LeCun, born July 8, 1960, 
                is a French computer scientist working primarily in the fields of machine learning, 
                computer vision, mobile robotics, and computational neuroscience. 
                He is the Silver Professor of the Courant Institute of Mathematical Sciences at New York University, and Vice President, Chief AI Scientist at Meta.
"""
```

```python
from nltk.tokenize import word_tokenize
```

```python

word_tokens = word_tokenize(sample_text)
  
filtered_text = [w for w in word_tokens if not w.lower() in stopwords.words('english')]
# print(word_tokens)
print(filtered_text)
```
